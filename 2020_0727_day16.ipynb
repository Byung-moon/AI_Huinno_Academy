```python
x = np.arange(0, 10)
y = np.arange(0,100,10)

#shorthand notation
plt.plot(x,y, 'g-.D', label='10x')
plt.plot(x, x**3, 'r:*', label ='^3')
plt.title("Second Graph")

plt.xticks(x)
#plt.yticks(y)

plt.legend()
plt.show()
```

```python
x = np.arange(0, 10)
y = np.arange(0,100,10)

#shorthand notation
plt.plot(x,y, 'g-.D', label='10x')
plt.plot(x[:4], x[:4]**3, 'r:*', label ='^3')
plt.plot(x[3:], x[3:]**3, 'b-o')
plt.title("Second Graph")

plt.xticks(x)
#plt.yticks(y)

plt.legend()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure(figsize = (9,9), dpi=150)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])
plt.suptitle('Fig. 1')

plt.figure(2)
plt.plot([4,5,6,7])
plt.grid(True)
plt.title('Figure 2.')

plt.figure(3)
plt.subplot(121)
plt.bar([1,2,3], [3,6,9])
plt.title('Fig. 3(a)')
plt.subplot(122)
plt.scatter([1,2,3],[3,6,9])
plt.title('Fig. 3(b)')
plt.suptitle('Fig. 3. ~~~~~~')


plt.figure(1)
plt.subplot(211)
plt.title('Fig. 1(a)')

plt.show()
```
